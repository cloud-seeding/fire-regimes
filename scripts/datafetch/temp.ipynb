```python
from netCDF4 import Dataset

# Open one of your .nc files
file_path = './assets/all/all_2.0_20040104.nc'
with Dataset(file_path, 'r') as nc:
    # Print global attributes
    print("Global Attributes:")
    for attr in nc.ncattrs():
        print(f"\t{attr}: {getattr(nc, attr)}")

    # Print dimensions
    print("\nDimensions:")
    for dim in nc.dimensions.values():
        print(f"\t{dim.name}: {dim.size}")

    # Print variables
    print("\nVariables:")
    for var in nc.variables.values():
        print(f"\t{var.name}: {var.dimensions}")
        print(f"\t\tShape: {var.shape}")
        print(f"\t\tDatatype: {var.dtype}")
        for attr in var.ncattrs():
            print(f"\t\t{attr}: {getattr(var, attr)}")
        print()
```
